```python
import pandas as pd
```

```python
df = pd.read_csv('003 salaries-by-college-major.csv')
```

```python
clean_df = df.dropna()
```

```python
max_salary = clean_df['Starting Median Salary'].idxmax()
```

```python
clean_df.loc[max_salary]
```

```python
highest_mid_salary = clean_df['Mid-Career Median Salary'].idxmax()
```

```python
clean_df['Undergraduate Major'][highest_mid_salary]
```

```python
clean_df['Mid-Career Median Salary'][highest_mid_salary]
```

```python
lowest_starting_salary = clean_df['Starting Median Salary'].idxmin()
```

```python
clean_df['Undergraduate Major'][lowest_starting_salary]
```

```python
lowest_mid_salary = clean_df['Mid-Career Median Salary'].idxmin()
```

```python
clean_df['Undergraduate Major'][lowest_mid_salary]
```

```python
clean_df['Mid-Career Median Salary'][lowest_mid_salary]
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
```

```python
low_risk[['Spread', 'Undergraduate Major']].head()
```

```python
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
```

```python
highest_potential[['Mid-Career 90th Percentile Salary', 'Undergraduate Major']].head()
```

```python
greatest_spread = clean_df.sort_values('Spread', ascending=False)
```

```python
greatest_spread[['Spread', 'Undergraduate Major']].head()
```
